```python
df = df.drop(['Reason for Absence'], axis = 1)
df
```

```python
# We want to group reasons 1-14, 15-17, 18-21, and 22+
reason_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
# Check one out 
reason_1
```

```python
# Concactenating columns 
# Merge our new dummies with the data frame 
df = pd.concat([df, reason_1, reason_2, reason_3, reason_4], axis = 1)
df
```

```python
# Assign column names 
# First pull them out to copy and paste, then add new ones
df.columns.values
```

```python
column_names = ['ID', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = column_names
df.head()
```

```python
# Re-order the columns 
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'ID', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
df.head()
```

```python
# Create a checkpoint (similar concept to commit and rollback in SQL)
# Create a copy of the new data frame 
df_reason_modify = df.copy()
```

```python
# Run: df_reason_modify to come back to this version of our data frame 
```

```python
# Convert date values into a time stamp 
# Syntax: pd.to_datetime(dataframe[column],format=)
df_reason_modify['Date'] = pd.to_datetime(df_reason_modify['Date'])
df_reason_modify['Date']
```

```python
# Lets extract the months 
df_reason_modify['Date'][0].month
# Obs 1 (presented as 0) is July 
```

```python
# Create empy data frame 
list_months = []
list_months
```

```python
df_reason_modify.shape
```

```python
for i in range(df_reason_modify.shape[0]):
    list_months.append(df_reason_modify['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
# Add this as a new column 
df_reason_modify['Month Value'] = list_months
df_reason_modify.head(20)
```

```python
# Doing the same but with weekdays 
# Reminder: Monday = 0 
df_reason_modify['Date'][0].weekday()
# Obs 1 (presented as 0) is Tuesday
```

```python
# Create a function 
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
# Apply the function 
df_reason_modify['Day of Week']= df_reason_modify['Date'].apply(date_to_weekday)
df_reason_modify.head(20)
```

```python
# Transform education into a dummy variable 
df_reason_modify['Education'].unique()
```

```python
# 1 = highschool, 2 = graduate, 3 = postgraduate, 4 = master / doctorate 
# Check how many times each level is encountered
df_reason_modify['Education'].value_counts()
```

```python
# Combine the last 3 levels of education for the dummy variable 
df_reason_modify['Education'] = df_reason_modify['Education'].map({1:0, 2:1, 3:1, 4:1})
# The number of keys needs to be the same as the number of values 
```
